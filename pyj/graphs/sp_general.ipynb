# subplots in Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
mtcars = data('mtcars')
mtcars.head()
mtcars = mtcars.sort_values(by='wt')
```

```python
fig, axs = plt.subplots(2,3, sharex='col', sharey='row')
plt.show()
#2Ã—3 grid of subplots, where all axes in the same row share their y-axis scale, 
# and all axes in the same column share their x-axis scale:
```

```python
# basic subplot
# wt vs mpg
fig, axs = plt.subplots()
axs.plot(mtcars.wt, mtcars.mpg)
plt.show()
#wt should be ordered
```

```python
# multiple plots - index
# wt vs mpg, hp
fig, axs = plt.subplots(nrows=2, gridspec_kw={'hspace':.4})
axs[0].plot(mtcars.wt, mtcars.mpg)
axs[1].plot(mtcars.wt, mtcars.hp)
plt.show();
```

```python
# multiple plots - naming
# wt vs mpg, hp
fig, (ax1, ax2,ax3) = plt.subplots(ncols=3, figsize=(10,6), width_ratios=[.2, .3, .5], gridspec_kw={'wspace':.2})
ax1.plot(mtcars.wt, mtcars.mpg)
ax2.plot(mtcars.wt, mtcars.hp)
ax3.plot(mtcars.wt, mtcars.qsec)
plt.show();
```

```python
# multiple plots - naming
# wt vs mpg, hp
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,6), sharex=True, height_ratios=[.4,.6])
#nrows > 1 if ncols>=2
axs[0,0].plot(mtcars.wt, mtcars.mpg, 'tab:orange')
axs[0,0].set_title('wt vs mp')
axs[0,1].plot(mtcars.wt, mtcars.hp,'red')
axs[0,1].set_title('wt vs hp')
axs[1,0].plot(mtcars.wt, mtcars.qsec)
axs[1,0].set_title('wt vs qsec')
axs[1,1].plot(mtcars.wt, mtcars.drat)
axs[1,1].set_title('wt vs drat')
fig.suptitle('wt vs mpg, hp, qsec, drat')
fig.tight_layout()
plt.show();
```

```python
# label values
fig, axs = plt.subplots(figsize=(10,8))
axs.plot(mtcars.wt, mtcars.mpg)
for index, row in mtcars.iloc[0:31].iterrows():
    axs.text(x=row['wt'], y=row['mpg'], s=index)
axs.set_xlabel('Weight')
axs.set_ylabel('Mileage')
fig.suptitle('Wt vs Mpg with labels')
plt.show()
```

# complicated arrangements
- To go beyond a regular grid to subplots that span multiple rows and columns, plt.GridSpec() is the best tool. The plt.GridSpec() object does not create a plot by itself; it is simply a convenient interface that is recognized by the plt.subplot() command. For example, a gridspec for a grid of two rows and three columns with some specified width and height space looks like this:
- https://jakevdp.github.io/PythonDataScienceHandbook/04.08-multiple-subplots.html

```python
grid = plt.GridSpec(2,3, wspace=.4, hspace=.3)
```

```python
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
```

```python
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])  #spans 2 cols
plt.subplot(grid[1, :2]) #spans 2 cols
plt.subplot(grid[1, 2]) 
plt.show()
```
